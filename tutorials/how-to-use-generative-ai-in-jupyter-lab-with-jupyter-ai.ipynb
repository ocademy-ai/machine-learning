## Generate Python code

````python
%%ai chatgpt -f code

Generate Python code for inserting sort.
````

````python
def insertionSort(arr):
    for i in range(1, len(arr)):
        key = arr[i]
        j = i-1
        while j >=0 and key < arr[j]:
            arr[j+1] = arr[j]
            j -= 1
        arr[j+1] = key

arr = [64, 34, 25, 12, 22, 11, 90]
insertionSort(arr)
print("Sorted array is:")
for i in range(len(arr)):
    print("%d" % arr[i], end=" ")
````

### Interact with Jupyternaut

Now, let's try to rewrite the code with comments added to it by using Jupyternaut.



Then you will get below code with explanation as such text.

> The code defines a function called insertionSort that takes an array arr as input. It implements the insertion sort algorithm, which sorts the array in ascending order.
> The algorithm works by iterating over the array starting from the second element (i = 1). For each element, it compares it with the elements before it and shifts elements greater than the key to the right until it finds the correct position for the key.
> The key element is stored in the variable key. The variable j is used to keep track of the position being compared with the key.
> The while loop continues as long as j is greater than or equal to 0 and the key is less than the element at index j. This loop compares the key with the elements before it and shifts them to the right if necessary.
> Inside the loop, the element at index j is moved to the right by assigning arr[j+1] the value of arr[j]. This effectively shifts the element one position to the right.
> After the loop, the key is placed at its correct position by assigning arr[j+1] the value of key.
> I hope this helps! Let me know if you have any further questions.

````python
def insertionSort(arr):
    for i in range(1, len(arr)):  # Iterate over the array starting from the second element
        key = arr[i]  # Store the current element as the key
        j = i-1  # Start comparing with the element before the current element

        while j >= 0 and key < arr[j]:  # Compare the key with elements before it
            arr[j+1] = arr[j]  # Shift elements greater than the key to the right
            j -= 1  # Decrement j to move to the previous element

        arr[j+1] = key  # Place the key at its correct position
````

````python
import unittest

class InsertionSortTest(unittest.TestCase):

    def test_insertion_sort(self):
        arr = [64, 34, 25, 12, 22, 11, 90]
        insertionSort(arr)
        self.assertEqual(arr, [11, 12, 22, 25, 34, 64, 90])

    def test_insertion_sort_with_empty_array(self):
        arr = []
        insertionSort(arr)
        self.assertEqual(arr, [])

    def test_insertion_sort_with_single_element(self):
        arr = [1]
        insertionSort(arr)
        self.assertEqual(arr, [1])

    def test_insertion_sort_with_sorted_array(self):
        arr = [10, 20, 30, 40, 50]
        insertionSort(arr)
        self.assertEqual(arr, [10, 20, 30, 40, 50])

    def test_insertion_sort_with_reversed_array(self):
        arr = [50, 40, 30, 20, 10]
        insertionSort(arr)
        self.assertEqual(arr, [10, 20, 30, 40, 50])

unittest.main(argv=[''], verbosity=2, exit=False)
````

### Generate matplotlib sample code

````python
%%ai chatgpt -f code

Generate a program of Python about how to draw a normal distribution plot.
````

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

mu = 100  # mean
sigma = 15  # standard deviation
num_samples = 1000

samples = np.random.normal(mu, sigma, num_samples)

plt.hist(samples, bins=30, density=True, alpha=0.5, color='b')

x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, 1/(sigma*np.sqrt(2*np.pi)) * np.exp(-(x-mu)**2 / (2*sigma**2)), color='r')

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Normal Distribution')
plt.grid()

plt.show()
````
